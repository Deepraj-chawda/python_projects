```python
import requests
import bs4
```

```python
url = 'https://grras.com/course/python-programming-training-in-indore-jaipur-ahmedabad'
```

```python
resp = requests.get(url)
```

```python
resp.status_code
```

```python
soup = bs4.BeautifulSoup(resp.content,'html')
```

```python
soup
```

```python
type(soup)
```

```python
data = {
    'users' : [],
    'reviews' : [],
    'time' : [],
    'lab_infra' : [],
    'training_delivery' : [],
    'Guidance' : []
 }
```

```python
divs = soup.find_all('div',attrs={'class':'review-wrap-content'})
```

```python
for div in divs:
    review = div.p.text
    if not review.strip():
        review = div.div.text
    data['reviews'].append(review)
```

```python
for div in divs:
    name, time = div.h3.text.split('(')
    data['users'].append(name)
    data['time'].append(time[:-1])
```

```python
for div in divs:
    rating = []
    for d in div.find_all('div',attrs={'class':'row review-feature-wrapper'})[0].find_all('div',attrs={'class','col-lg-3 col-md-6 review-feature'}):
        rating.append(len(d.find_all('span',attrs={"class":'fa fa-star'})))
    data['training_delivery'].append(rating[0])
    data['lab_infra'].append(rating[1])
    data['Guidance'].append(rating[2])
```

```python
for name,time,review,t_r,l_r,g_r in \
zip(data['users'],data['time'],data['reviews'],data['training_delivery'],data['lab_infra'],data['Guidance']):
    print("Name : ",name)
    print(time)
    print('review : ',review)
    print(f'Rating : \nTraining Delivery : {t_r}\tLab Infrastructure : {l_r}\tGuidance : {g_r}')
    print('__'*30,'\n\n')
```

```python
import pandas as pd
```

```python
all_data = pd.DataFrame(data)
```
