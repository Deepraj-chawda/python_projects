```python
import pandas as pd
import requests
import bs4
```

```python
def review_stars(soup_obj):
    try:
        revi_st=soup_obj.find('a',{"class":"a-link-normal"}).get_text().strip()
        return revi_st
    except Exception as e:
        print(e)
```

```python
def review_comments(soup_obj):
    try:
        comm_rev=soup_obj.find('span',{"class":"a-size-base review-text review-text-content"}).get_text().strip()
        return comm_rev
    except Exception as e:
        print(e)
```

```python
def review_size(soup_obj):
    try:
        comm_rat=soup_obj.find('div',{"class":"a-row a-spacing-mini review-data review-format-strip"}).get_text().strip()
        return comm_rat
    except Exception as e:
        print(e)
```

```python
def review_date(soup_obj):
    try:
        comm_dat=soup_obj.find('span',{"class":"a-size-base a-color-secondary review-date"}).get_text().strip()
        return comm_dat
    except Exception as e:
        print(e)
```

```python
def review_name(soup_obj):
    try:
        comm_nam=soup_obj.find('div',{"class":"a-row a-spacing-mini"}).get_text().strip()
        return comm_nam
    except Exception as e:
        print(e)
```

```python
def get_review_component(soup_obj):
    return   {
             'custmer name'  :     review_name(soup_obj),
             'stars'        :    review_stars(soup_obj),
              'comments'       :    review_comments(soup_obj),
              'date'          :     review_date(soup_obj),
              'size'          :     review_size(soup_obj)
           }
```

```python
def amazon_reviews(url,header,no_of_page=1):
    s1 = pd.DataFrame({
             'custmer name'  :    [],
             'stars'        :    [],
              'comments'       :   [],
              'date'          :     [],
              'size'          :     []
           })
    for i in range(1,no_of_page+1):
        resp=requests.get(url.format(i,i),headers=header)
        soup=bs4.BeautifulSoup(resp.text,"html.parser")

        review=soup.find_all("div",{"class":"a-row a-spacing-none"})
        records=[get_review_component(rev) for rev in review]
    
  
        df=pd.DataFrame.from_records(records)
        s1 = pd.concat([s1, df], ignore_index=True)
        
    return s1
```
